```python
import os

path = './crossCulturalDeception.2014/EnglishUS/'

data = []
for name in ['ab', 'bf', 'dp']:
    for fp in os.listdir(os.path.join(path, name)):
        f_name = os.path.join(path, name, fp)
        try:
            with open(f_name, 'r', encoding='utf-8') as f:
                lines = f.readlines()
                for l in lines:
                    l = l.strip().split('\t')
                    label = 1 if 'False' in l[0] else 0
                    l = {'text':l[1], 'label':label}
                    data.append(l)
        except:
            pass
                
```

```python
print(len(data))
```

```python
import random

random.shuffle(data)

train = data[:360]

test = data[360:]
```

```python
f_data_ = open('./crossCulture_/test_x.csv', 'w', encoding='utf-8')
f_label_ = open('./crossCulture_/test_y.csv', 'w', encoding='utf-8')

for line in test:
    f_data_.write(line['text']+'\n')
    f_label_.write(str(line['label'])+'\n')
```

```python
f_data = open('./crossCulture_/train_x.csv', 'w', encoding='utf-8')
f_label = open('./crossCulture_/train_y.csv', 'w', encoding='utf-8')

for line in train:
    print(line)
    f_data.write(line['text']+'\n')
    f_label.write(str(line['label'])+'\n')
```
